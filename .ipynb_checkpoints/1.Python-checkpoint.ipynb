```python
def sq(x):
    return x*x

def mymap(a, f):
    retval = []
    for i in a:
        v = f(i)
        retval.append(v)
    return retval
```

```python
class PartyAnimal:
    
    x = 0
    name = ""
    
    def __init__(self, name):
        self.name = name
        print ("I am constructor..")
        
    def party(self):
        self.x = self.x + 1
        print (self.name," So far ", self.x)
    
class FootballFan(PartyAnimal):
    
    points = 0
    def touchdown(self):
        self.points = self.points + 7
        self.party()
        print (self.name, "Points ", self.points)
        
s = PartyAnimal("Jit")
s.party()
j = FootballFan("Jitu")
j.touchdown()
```

```python
def multiply_loops(A, B):
    c = np.zeros((A.shape[0], B.shape[1]))
    for i in range(A.shape[1]):
        for j in range(B.shape[0]):
            c[i, j] = A[i, j] * B[j, i]
            
    return c
```

```python
# Matrix Multiplicatoin using numpy
def multiply_vector(A, B):
    return A @ B
```

```python
import numpy as np

x = np.random.random((1000, 1000))
y = np.random.random((1000, 1000))
```

```python
%timeit multiply_loops(x, y)
```

```python
%timeit multiply_vector(x, y)
```
